```python
def sigmoid(z):
    z = np.clip(z, -500, 500)
    g = 1.0/(1.0+np.exp(-z))
    return g
```

```python
def compute_cost_logistic(X, y, w, b):
    m = X.shape[0]
    z = np.dot(X, w) + b
    f = sigmoid(z)
    cost = y*np.log(f) + (1 - y)*np.log(1 - f)
    cost = np.sum(cost)/m
    return cost
    
```

```python
w_tmp = np.array([1,1])
b_tmp = -3
```

```python
print(compute_cost_logistic(X_train, y_train, w_tmp, b_tmp))
```
